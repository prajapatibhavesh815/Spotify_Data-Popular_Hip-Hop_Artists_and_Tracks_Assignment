```python
#1. Identify the top 5 popular artists based on the mean popularity of their tracks.Show the mean popularity of tracks for the top 5 popular artists Using BarPlot.

# Group by artists and calculates the mean popularity of tracksfor each artist
artist_mean_popularity = df.groupby('Artist')['Popularity'].mean()

#Sort artists based on mean popularity in descending order and select the top 5
top_artists = artist_mean_popularity.sort_values(ascending=False).head(5)

#plotting the mean popularity of tracks for the top 5 popular artists
plt.figure(figsize=(10,6))
top_artists.plot(kind='bar',color='skyblue')
plt.title('mean popularity of tracks for top 5 popular artists')
plt.xlabel('Artists')
plt.ylabel('Mean popularity')
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.show()
```

```python
#2. Determine the top 5 popular songs based on their popularity ratings.Display the popularity ratings of the top 5 popular songs using BarPlot.


# Sort tracks based on popularity ratings in descending order and select the top 5
top_tracks = df.sort_values(by='Popularity', ascending=False).head(5)

# plotting the popularity ratings in descending order and select the top 5
plt.figure(figsize=(12, 8))
plt.bar(top_tracks['Track Name'], top_tracks['Popularity'], color='skyblue')
plt.title('Popularity Ratings of Top 5 Popular Tracks')
plt.xlabel('Track Name')
plt.ylabel('Popularity Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#4. Identify the top 5 longest songs among the tracks of the top 5 popular artists.Represent the duration of the top 5 longest songs among the tracks of the top 5 popular artists using BarPlot.

# Assuming df is your DataFrame containing track data
# and it has columns 'artist_name' and 'track_duration'



# Step 1: Sort tracks based on duration in descending order
top_longest_songs = df.groupby('Artist').apply(lambda x: x.nlargest(5, 'Duration (ms)'))

# Step 2: Select the top 5 longest songs
top_5_longest_songs = top_longest_songs.head(5)

# Step 3: Plotting the duration of the top 5 longest songs
plt.figure(figsize=(10, 6))
plt.bar(top_5_longest_songs['Artist'], top_5_longest_songs['Duration (ms)'], color='skyblue')
plt.title('Duration of Top 5 Longest Songs Among Tracks of Top 5 Popular Artists')
plt.xlabel('Song')
plt.ylabel('Duration (seconds)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#5. Determine the top 5 most danceable songs based on their danceability scores.Illustrate the danceability scores of the top 5 most danceable songs using PieChart.

# Assuming df is your DataFrame containing song data
# and it has columns 'song_name' and 'danceability_score'

# Step 1: Sort songs based on danceability scores in descending order
top_danceable_songs = df.sort_values(by='Artist', ascending=False)

# Step 2: Select the top 5 most danceable songs
top_5_danceable_songs = top_danceable_songs.head(5)

# Step 3: Illustrate the danceability scores of the top 5 most danceable songs using a pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_5_danceable_songs['Popularity'], labels=top_5_danceable_songs['Track Name'], autopct='%1.1f%%', startangle=140)
plt.title('Danceability Scores of Top 5 Most Danceable Songs')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
